```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r"D:\IBM data science\final project\train.csv")
sns.histplot(data['Item_Outlet_Sales'], bins=30, kde=True)
plt.title('Sales Distribution')
plt.show()
```

Visualizations:

1-Sales Distribution:

Insight: The histogram shows that most sales values are clustered at the lower end with fewer occurrences at the higher end.

Analysis: This right-skewed distribution indicates that a majority of items have lower sales, while a smaller number of items account for significantly higher sales. This is common in retail sales data where a few products drive the majority of the sales.

```python
sns.lineplot(data=data, x='Outlet_Establishment_Year', y='Item_Outlet_Sales')
plt.title('Sales Trends Over Years')
plt.show()
```

2- Sales Trends Over Years:

Insight: The line graph shows the trend of item outlet sales over the years, with a noticeable drop around the year 2000 and fluctuations in sales in the subsequent years.

Analysis: This trend suggests that sales vary significantly depending on the year the outlet was established. The drop around 2000 might be due to market conditions, changes in consumer behavior, or other external factors.

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = data.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

3- Correlation Analysis:

Insight: Item_MRP has a moderate positive correlation with sales, while Item_Visibility shows a slight negative correlation.

Analysis: Higher maximum retail prices (MRP) tend to correlate with higher sales, suggesting a premium pricing strategy might work well. The negative correlation between visibility and sales might imply that overexposed items aren't performing as well, possibly due to market saturation or consumer preferences.

The image is a correlation heatmap showing the following key points:

Item_MRP and Item_Outlet_Sales have a correlation coefficient of 0.57, indicating a positive relationship.

Item_Visibility and Item_Outlet_Sales have a correlation coefficient of -0.13, indicating a slight negative relationship.

Other correlations are relatively weak, suggesting less impactful relationships between those variables.
