```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r"D:\IBM data science\final project\train.csv")
```

```python
# Display basic info about the dataset
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
# Sales Distribution
sns.histplot(data['Item_Outlet_Sales'], bins=30, kde=True)
plt.title('Sales Distribution')
plt.show()
```

```python
# Sales Trends Over Years
sns.lineplot(data=data, x='Outlet_Establishment_Year', y='Item_Outlet_Sales')
plt.title('Sales Trends Over Years')
plt.show()
```

```python
numeric_cols = data.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Box Plot for Item MRP
sns.boxplot(data=data, x='Item_MRP')
plt.title('Box Plot of Item MRP')
plt.show()
```

```python
#sales by outlet_Type
sns.boxplot(data=data, y='Item_Outlet_Sales', x='Outlet_Type')
plt.title('Sales by Outlet Type')
plt.show()
```

```python
sns.boxplot(data=data, y='Item_Outlet_Sales', x='Item_Fat_Content')
plt.title('Sales by Item Fat Content')
plt.show()
```

```python
sns.boxplot(data=data, y='Item_Outlet_Sales', x='Item_MRP')
plt.title('Sales by Item MRP')
plt.show()
```

```python
sns.boxplot(data=data, y='Item_Outlet_Sales', x='Outlet_Size')
plt.title('Sales by Outlet Size')
plt.show()
```
